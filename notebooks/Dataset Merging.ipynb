```python
import pandas as pd
import numpy as np
import os

features_df = pd.read_csv('C:\\Users\\gmoor\\Documents\\REVISED CAPSTONE PROJECT\\Data\\new_sprectal_indices.csv')
```

```python
onehundred_features_df = features_df[features_df['segment_type'] == '100m']

onehundred_features_df
```

```python
import geopandas as gpd

onehundred_segment_gdf = gpd.read_file('C:\\Users\\gmoor\\Documents\\REVISED CAPSTONE PROJECT\\Data\\first200_aggregate_segments (100meters).gpkg')
```

```python
onehundred_segment_gdf
```

```python
# Create duplicates with different buffer sizes
buffer_sizes = [100, 200, 300]
duplicated_dfs = []

for buffer_size in buffer_sizes:
    temp_df = onehundred_segment_gdf.copy()
    temp_df['buffer_size'] = buffer_size
    duplicated_dfs.append(temp_df)

# Concatenate all the duplicated dataframes
duplicated_onehundred_segment_gdf = pd.concat(duplicated_dfs, ignore_index=True)
```

```python
duplicated_onehundred_segment_gdf[(duplicated_onehundred_segment_gdf['Year'] == 2014) & (duplicated_onehundred_segment_gdf['100m_segment_id'] == 1)]
```

```python
onehundred_segment_combined_df = pd.merge(
    duplicated_onehundred_segment_gdf,
    onehundred_features_df,
    left_on=['100m_segment_id', 'Year', 'buffer_size'],
    right_on=['segment_id', 'year', 'buffer_size']
)

onehundred_segment_combined_df.head()
```

```python
onehundred_segment_combined_df[(onehundred_segment_combined_df['Year'] == 2014) & (onehundred_segment_combined_df['100m_segment_id'] == 1)]
```

```python
twohundred_features_df = features_df[features_df['segment_type'] == '200m']

twohundred_features_df
```

```python
import geopandas as gpd

twohundred_segment_gdf = gpd.read_file('C:\\Users\\gmoor\\Documents\\REVISED CAPSTONE PROJECT\\Data\\first80_aggregate_segments (200meters).gpkg')
```

```python
twohundred_segment_gdf
```

```python
# Create duplicates with different buffer sizes
buffer_sizes = [200, 300, 400]
duplicated_dfs = []

for buffer_size in buffer_sizes:
    temp_df = twohundred_segment_gdf.copy()
    temp_df['buffer_size'] = buffer_size
    duplicated_dfs.append(temp_df)

# Concatenate all the duplicated dataframes
duplicated_twohundred_segment_gdf = pd.concat(duplicated_dfs, ignore_index=True)
```

```python
duplicated_twohundred_segment_gdf[(duplicated_twohundred_segment_gdf['Year'] == 2007) & (duplicated_twohundred_segment_gdf['200m_segment_id'] == 1)]
```

```python
twohundred_segment_combined_df = pd.merge(
    duplicated_twohundred_segment_gdf,
    twohundred_features_df,
    left_on=['200m_segment_id', 'Year', 'buffer_size'],
    right_on=['segment_id', 'year', 'buffer_size']
)

twohundred_segment_combined_df.head()
```

```python
twohundred_segment_combined_df.describe()
```

```python
twohundred_segment_combined_df[(twohundred_segment_combined_df['Year'] == 2014) & (twohundred_segment_combined_df['200m_segment_id'] == 1)]
```

```python
total_segment_combined_df = pd.concat([onehundred_segment_combined_df, twohundred_segment_combined_df], ignore_index=True)

total_segment_combined_df.head()
```
